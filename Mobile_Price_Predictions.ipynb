# **Standard Scaler**

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
X=data_train.drop('price_range', axis=1)
```

```python
X
```

```python
Y=data_train['price_range']
Y
```

```python
X_std=std.fit_transform(X)
data_test_std=std.transform(data_test)
X_std
```

## **MODEL**

# **1.Decision Tree**


```python
!pip install catboost
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
import xgboost as xgb
from catboost import CatBoostClassifier

dt = DecisionTreeClassifier()
dt.fit(X_std, Y)
dt_predictions = dt.predict(data_test_std)
dt_score = dt.score(X_std, Y)
print(dt_predictions)
print(f"Decision Tree Accuracy: {dt_score}")
```

AdaBoost

```python
ada = AdaBoostClassifier(estimator=DecisionTreeClassifier(), n_estimators=50)
ada.fit(X_std, Y)
ada_predictions = ada.predict(data_test_std)
ada_score = ada.score(X_std, Y)
print(dt_predictions)
print(f"AdaBoost Accuracy: {ada_score}")
```

GradientBoost

```python
gb = GradientBoostingClassifier(n_estimators=100)
gb.fit(X_std, Y)
gb_predictions = gb.predict(data_test_std)
gb_score = gb.score(X_std, Y)
print(gb_predictions)
print(f"Gradient Boosting Accuracy: {gb_score}")
```

XGBoost

```python
xg = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
xg.fit(X_std, Y)
xg_predictions = xg.predict(data_test_std)
xg_score = xg.score(X_std, Y)
print(xg_predictions)
print(f"XG Boosting Accuracy: {gb_score}")
```

CatBoost

```python
cat = CatBoostClassifier(iterations=100, verbose=0)
cat.fit(X_std, Y)
cat_predictions = cat.predict(data_test_std)
cat_score = cat.score(X_std, Y)
print(cat_predictions)
print(f"CatBoost Accuracy: {cat_score}")
```
